```python
from scipy.stats import chi2_contingency
  
# defining the table
d = data[['PCC1', 'DCC1']]
stat, p, dof, expected = chi2_contingency(d)
  
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

```python
# chi-squared test with similar proportions
from scipy.stats import chi2_contingency
from scipy.stats import chi2

stat, p, dof, expected = chi2_contingency(d)
print('dof=%d' % dof)
print(expected)
# interpret test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
if abs(stat) >= critical:
	print('Dependent (reject H0)')
else:
	print('Independent (fail to reject H0)')
# interpret p-value
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
	print('Dependent (reject H0)')
else:
	print('Independent (fail to reject H0)')
```

### KModes

```python
from kmodes.kmodes import KModes

kmodes = KModes(n_clusters=2, random_state=0) 

result = kmodes.fit(data)

result
```

```python
# Building the model with 3 clusters
kmodes = KModes(n_clusters=3, init = "random", n_init = 5, verbose=1)
clusters = kmodes.fit_predict(data)
clusters
```

```python
# Elbow curve to find optimal K
cost = []
K = range(1,5)
for num_clusters in list(K):
    kmodes = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=1)
    kmodes.fit_predict(data)
    cost.append(kmodes.cost_)
    
plt.figure(figsize=(10,8))
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters', fontsize=14)
plt.ylabel('Cost',fontsize=14)
plt.title('Elbow Method For Optimal k',fontsize=14)
plt.grid()
#plt.savefig("Elbow.png", dpi=300)
plt.show()
```

### NMF

```python
from sklearn.decomposition import NMF

#create model
nmf_model = NMF(n_components=3, 
                init='nndsvda', 
                solver = 'mu',
                beta_loss = 'kullback-leibler',
                l1_ratio = 0.5,
                random_state = 101)
nmf_model.fit(data)
```

```python
def predict(w_matrix):
    sortedW = np.argsort(w_matrix)
    n_predictions, maxValue = sortedW.shape
    predictions = [[sortedW[i][maxValue - 1]] for i in range(n_predictions)]
    topics = np.empty(n_predictions, dtype = np.int64)
    for i in range(n_predictions):
        topics[i] = predictions[i][0]
    return topics
```

```python
cluster = predict(nmf_model.transform(data))

cluster
```

```python
data["types_NMF"] = cluster
```
